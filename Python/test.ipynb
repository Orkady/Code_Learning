```python
'''
编写一个程序，求出不大于给定正整数n的全部质数。
质数是指只能被1和它本身整除的大于1的自然数，例如：2, 3, 5, 7, 11等。
程序需要从2开始，找出所有符合条件的质数，并以空格分隔的方式打印出来。

平台会对你编写的代码进行测试：

测试输入：10；
预期输出：2 3 5 7

测试输入:20；
预期输出：2 3 5 7 11 13 17 19
'''
n = input()

# 代码开始
n = int(n)
# 遍历从2到n的所有数字，寻找质数
for i in range(2, n+1):
    # 检查当前数字i是否为质数
    for j in range(2, i):
        # 如果i能被j整除，则i不是质数，跳出内层循环
        if i % j == 0:
            break
    # 如果内层循环正常结束（没有找到因数），则i是质数，打印出来
    else:
        print(i, end=" ")
# 代码结束
```

上方有个特殊的for-else用法，
意思为只有当for循环正常结束，才会执行else中的代码。
如果因为break 而提前结束，则不会执行。


```python
'''
编写一个程序，统计在给定范围[L, R]内所有整数中数字3出现的次数。
例如，对于范围[3, 33]，数字3在数3中出现1次，在数13中出现1次，在数23中出现1次，在数30中出现1次，在数31中出现1次，在数32中出现1次，在数33中出现2次，因此数字3在该范围内一共出现了8次。


'''

s = input().split()

# 代码开始
# 从字符串s中提取前两个字符作为数字范围的左右边界
L = int(s[0])
R = int(s[1])
# 初始化计数器，用于统计数字3出现的次数
count = 0

# 遍历从L到R的每个数字（包含R）
for i in range(L, R + 1):
    # 将当前数字转换为字符串，便于逐位检查
    i = str(i)
    # 遍历当前数字的每一位字符
    for j in i:
        # 如果当前位是字符"3"，则计数器加1
        if j == "3":
            count += 1
        else:
            # 如果不是"3"则跳过当前循环
            continue

# 输出统计结果
print(count)
# 代码结束
```

```python
'''
编写一个程序，通过输入函数获取一个十进制数值，并将其转换为二进制数输出到屏幕。
转换的原理是：用2整除十进制整数，可以得到一个商和余数；
再用2去除商，又会得到一个商和余数，如此进行，直到商小于2时为止，然后把先得到的余数作为二进制数的低位有效位，后得到的余数作为二进制数的高位有效位，依次排列起来。

注意：请按照题目要求完成这道题，不要使用bin()函数，否则这题为0分。
'''

n = input()

# 代码开始
# 将输入的数值转换为整数类型
n = int(n)
# 初始化余数列表，用于存储二进制转换过程中的余数
yu_list = []

# 通过不断除以2的方式将十进制数转换为二进制数
while n > 0:
    # 当n等于1时，直接添加到余数列表并退出循环
    if n == 1:
        yu_list.append(1)
        break
    else:
        # 计算商和余数，将余数添加到列表中，然后更新n为商
        shang, yu = n // 2, n % 2
        yu_list.append(yu)
        n = shang

# 将余数列表反转，得到正确的二进制表示顺序
yu_list.reverse()

# print(yu_list)
# 按顺序输出二进制数的每一位数字
for i in yu_list:
    print(i, end="")
# 代码结束
```

```python
'''
模拟数据压缩中的行程长度压缩方法。
行程长度压缩方法是，对于一个待压缩的字符串而言，依次记录每个字符及重复的次数。
例如，待压缩字符串为AAABBBBCBB，则压缩的结果是(A,3)(B,4)(C,1)(B,2)。
字符串不区分大小写，即所有小写字母均可视为相应的大写字母。

测试输入：GGhhGggKKkkDhHHlLlOo；
预期输出：(G,2)(H,2)(G,3)(K,4)(D,1)(H,3)(L,3)(O,2)

测试输入：GGhhGggKKgkDhHHlLlOo；
预期输出：(G,2)(H,2)(G,3)(K,2)(G,1)(K,1)(D,1)(H,3)(L,3)(O,2)
'''

s = input()

# 代码开始
# 将字符串转换为大写形式
s_up = s.upper()
# print(s_up)

# 初始化计数器
count = 1
# 遍历字符串，统计连续相同字符的个数
for i in range(len(s_up)-1):
    # 如果当前字符与下一个字符相同，则计数器加1
    if s_up[i] == s_up[i+1]:
        count += 1
    # 如果当前字符与下一个字符不同，则输出当前字符及其计数，并重置计数器
    else:
        print('(', s_up[i], ',', count, ')', sep='', end='')
        count = 1
    # 处理字符串最后一个字符的情况
    if i == len(s_up)-2:
        print('(', s_up[i+1], ',', count, ')', sep='', end='')
# 代码结束
```

```python
'''
任务描述
本关任务：编写一个程序，输入一行字符，统计并输出其中数字字符的个数，并且输出一个只包含这些数字字符的字符串。

分析
要完成这个任务，我们需要执行以下步骤：

1.读取用户输入的一行字符。
2.遍历输入的每个字符，检查它是否是数字字符。
3.统计数字字符的总数，并且将这些数字字符存储在一个字符串中。
4.输出数字字符的个数以及只包含数字字符的字符串。

相关知识
为了完成本关任务，你需要掌握：

1.基本的输入输出操作。
2.循环结构，用于遍历输入的每个字符。
3.条件判断，用于检查字符是否是数字字符。
4.字符串连接，用于构建只包含数字字符的字符串。

字符串判断
在Python中，字符串的判断操作通常涉及到检查字符串的内容、类型或者特定的条件。对于本任务，我们需要判断一个字符是否为数字字符。这可以通过使用字符串的isdigit()方法来实现。

isdigit()方法
isdigit()方法用于检查字符串是否只包含数字，如果是，则返回True；否则返回False。这个方法对于检测单个字符是否为数字也非常有用。

# 检查单个字符是否为数字
char = '5'
if char.isdigit():
    print("这是一个数字字符")
else:
    print("这不是一个数字字符")
在循环中，我们可以使用isdigit()方法来过滤出所有的数字字符。

字符串连接
字符串连接是指将两个或多个字符串拼接在一起形成一个新的字符串。在Python中，这可以通过加号+操作符或者join()方法来实现。

加号+操作符
使用加号+操作符可以将两个字符串相加，即连接它们。

# 使用加号连接字符串
str1 = "Hello, "
str2 = "World!"
result = str1 + str2
print(result)  # 输出：Hello, World!
在循环中，我们可以使用加号+操作符来逐步构建包含所有数字字符的字符串。

join()方法
join()方法用于将序列中的元素连接成一个字符串，元素之间用指定的分隔符分隔。

# 使用join方法连接列表中的字符串元素
elements = ["Hello", "World", "!"]
result = " ".join(elements)
print(result)  # 输出：Hello World !
在本任务中，我们不直接使用join()方法，因为我们是逐个字符地构建数字字符串，而不是从一个列表中连接元素。

编程要求
根据提示，在右侧编辑器补充代码，实现以下功能：

1.输入一行字符。
2.统计其中数字字符的个数，并将数字字符存储在一个字符串中。
3.输出数字字符的个数以及只包含数字字符的字符串。

测试说明
平台会对你编写的代码进行测试：

测试输入：hello123world456
预期输出：
数字字符个数：6
数字字符串：123456

测试输入：123abc456
预期输出：
数字字符个数：6
数字字符串：123456
'''

input_line = input()

# 代码开始
# 处理输入字符串，提取所有数字字符并计数
count = 0           # 数字字符计数器
digit_list = ""   # 存储所有数字字符组成的字符串
for char in input_line:
    # 判断当前字符是否为数字字符
    if char.isdigit():
        count += 1
        digit_list += char
print("数字字符个数：", count, sep="")
print("数字字符串：", digit_list, sep="")
# 代码结束
```

```python
'''
任务描述
本关任务：编写一个程序，输入一个字符串，统计该字符串中每个字符出现的次数，并以字典的形式输出结果。

分析
要完成这个任务，我们需要执行以下步骤：

1.读取用户输入的字符串。
2.初始化一个数据结构（如字典）来存储每个字符及其出现的次数。
3.遍历字符串中的每个字符。
4.对于每个字符，检查它是否已经在字典中，如果在，则增加其计数；如果不在，则添加到字典并设置计数为1。
5.输出包含所有字符及其出现次数的字典。

相关知识
为了完成本关任务，你需要掌握：

1.基本的输入输出操作。
2.循环结构，用于遍历字符串中的每个字符。
3.字典操作，用于存储字符及其出现次数。

字典操作
字典（Dictionary）是Python中的一种内置数据结构，它以键值对（key-value pairs）的形式存储数据。字典的键是唯一的，而值可以是任何数据类型。字典提供了快速的数据查找功能，其查找速度通常比列表快。

创建字典
字典可以用花括号{}或者dict()函数来创建。

# 使用花括号创建字典
my_dict = {'name': 'Kimi', 'age': 30}
# 使用dict()函数创建字典
my_dict = dict(name='Kimi', age=30)
访问字典中的值
通过键来访问字典中的值。

# 访问字典中的值
my_dict = {'name': 'Kimi', 'age': 30}
name = my_dict['name']
print(name)  # 输出：Kimi
添加和修改字典中的值
给字典添加新的键值对或者修改已有键的值。

my_dict = {'name': 'Kimi', 'age': 30}
# 添加新的键值对
my_dict['gender'] = 'Female'
# 修改已有键的值
my_dict['age'] = 31
删除字典中的值
可以使用del语句或者pop()方法来删除字典中的键值对。

my_dict = {'name': 'Kimi', 'age': 30}
# 使用del语句删除键值对
del my_dict['gender']
# 使用pop()方法删除键值对，并获取被删除的值
age = my_dict.pop('age')
print(age)  # 输出：31
遍历字典
遍历字典时，可以遍历它的键、值或者键值对。

my_dict = {'name': 'Kimi', 'age': 30}
# 遍历字典的键
for key in my_dict:
    print(key)
# 遍历字典的值
for value in my_dict.values():
    print(value)
# 遍历字典的键值对
for key, value in my_dict.items():
    print(f"{key}: {value}")
编程要求
根据提示，在右侧编辑器补充代码，实现以下功能：

1.输入一个字符串。
2.统计并存储每个字符出现的次数。
3.输出字符及其出现次数的字典。

测试说明
平台会对你编写的代码进行测试：

测试输入：hello,world.；
预期输出：{'h': 1, 'e': 1, 'l': 3, 'o': 2, ',': 1, 'w': 1, 'r': 1, 'd': 1, '.': 1}

测试输入：jsfwqjfsakfhalfdjfhsfjd；
预期输出：{'j': 4, 's': 3, 'f': 6, 'w': 1, 'q': 1, 'a': 2, 'k': 1, 'h': 2, 'l': 1, 'd': 2}
'''

s = input()

# 代码开始
# 创建一个空字典
char_dict = {}
# 遍历字符串中的每个字符
for char in s:
    if char in char_dict:       # 如果字符已经存在于字典中
        char_dict[char] += 1    # 增加字符的计数
    else:
        char_dict[char] = 1     # 添加字符到字典中并设置计数为1

# 打印字符及其出现次数的字典
print(char_dict)
# 代码结束
```

```python
'''
任务描述
本关任务：编写一个程序，为校园食宿预订系统计算食堂的总预订收入。系统提供m个菜，每个菜有固定的菜价和总份数。共有n个学生点餐，每个学生点3个菜。当点的菜售罄时，学生就买不到这个菜了。请根据学生预订记录，给出食堂总的预订收入。（选做题，不影响成绩）

分析
要完成这个任务，我们需要执行以下步骤：

1.读取输入，包括学生数量n和菜的数量m。
2.初始化一个字典来存储每个菜的信息，包括价格和总份数。
3.遍历每个菜，读取其编号、价格和总份数，并存储到字典中。
4.遍历每个学生的预订记录，读取其点的3个菜的编号。
5.对于每个预订的菜，检查是否还有剩余份数，如果有，则减少对应的份数，并增加总收入。
6.输出食堂的总预订收入。

输入描述
程序的输入包含三部分：

学生数量和菜的数量
1.第一行包含两个整数，分别表示学生数量n和菜的数量m，两个数字之间用空格分隔。
2.学生数量n表示有多少个学生将进行点餐。
3.菜的数量m表示食堂提供多少种不同的菜。

每个菜的信息
1.接下来的m行，每行包含三个数据，分别表示一个菜的编号、价格和总份数。
2.菜的编号是一个字符串，通常是一个数字，表示菜的编号。
3.价格是一个整数，表示这个菜的价格。
4.总份数是一个整数，表示这个菜总共可以被点多少次。

学生的预订记录
1.紧接着的n行，每行包含三个数据，分别表示一个学生点的三个菜的编号。
2.每个编号都是字符串类型，对应之前输入的菜的编号。
3.每个学生点餐的三个菜之间用空格分隔。

相关知识
为了完成本关任务，你需要掌握：

1.基本的输入输出操作。
2.数据结构，如字典，用于存储菜的信息。可以使用字典来存储每个菜的信息，其中键为菜的编号，值为一个列表，包含价格和总份数。
3.循环结构，用于处理学生的预订记录和计算收入。使用循环来遍历学生的预订记录，并更新每个菜的已售份数和计算收入。

编程要求
根据提示，在右侧编辑器补充代码，实现以下功能：

1.读取学生数量n和菜的数量m。
2.初始化字典存储每个菜的信息。
3.读取每个菜的编号、价格和总份数，并存储到字典中。
4.读取学生的预订记录，并更新每个菜的已售份数。
5.计算并输出食堂的总预订收入。

测试说明
平台会对你编写的代码进行测试：

测试输入：
2 3
1 10 1
2 8 2
3 4 5
1 2 3
1 2 3
预期输出：34

测试输入：
3 4
1 10 1
2 8 2
3 4 5
4 2 5
1 2 3
1 2 3
2 3 4
预期输出：40
'''

import copy

# 读取学生数量n和菜的数量m
n, m = map(int, input().split())
menu = {}

# 读取每个菜的编号、价格和总份数
for i in range(m):
    s = input().split()
    menu[int(s[0])] = [int(s[1]), int(s[2])]
    

menu_booked = copy.deepcopy(menu)
# 读取学生的预订记录，并更新每个菜的已售份数
for i in range(n):
    t = input().split()
    for j in t:
        # 如果菜品还有剩余份数，则减少一份
        if menu_booked[int(j)][1] > 0:
            menu_booked[int(j)][1] -= 1
        else:
            # 如果菜品已售完，则跳过
            continue


# 计算并输出食堂的总预订收入
total = 0

# 遍历所有菜品，计算已售出菜品的总收入
for i in menu_booked:
    # 菜品价格 * 已售份数 = 该菜品收入
    total += menu_booked[i][0] * (menu[i][1] - menu_booked[i][1])
    
print(total)
```

```python
'''
任务描述
本关任务：编写一个函数，接收一串明文密码字符串作为参数，返回一个包含四个数字的元组。这四个数字分别代表字符串的长度、大写字母的个数、小写字母的个数和数字的个数。

分析
要完成这个任务，我们需要执行以下步骤：

1.定义一个函数，接收一个字符串参数。
2.初始化四个计数器，分别用于统计字符串的长度、大写字母的个数、小写字母的个数和数字的个数。
3.遍历字符串中的每个字符，根据字符的类型更新相应的计数器。
4.将四个计数器的值作为元组返回。

相关知识
为了完成本关任务，你需要掌握：

1.基本的输入输出操作。
2.条件判断，用于检查字符的类型。
3.循环结构，用于遍历字符串中的每个字符。

条件判断
使用isupper()、islower()和isdigit()方法来判断字符是否为大写字母、小写字母或数字。

编程要求
根据提示，在右侧编辑器补充代码，实现以下功能：

1.定义一个函数，接收一个字符串参数。
2.初始化四个计数器。
3.遍历字符串中的每个字符，并更新计数器。
4.返回包含四个计数器值的元组。

测试说明
平台会对你编写的代码进行测试：

测试输入：ABCabcd1234567；
预期输出：(14, 3, 4, 7)
'''

s = input()
# ------------------------------
def get_password(pwd):
    length = len(pwd)
    upper_count = sum(1 for char in pwd if char.isupper())
    lower_count = sum(1 for char in pwd if char.islower())
    digit_count = sum(1 for char in pwd if char.isdigit())
    return length, upper_count, lower_count, digit_count


print(get_password(s))
# -------------------------------
```

```python
'''
任务描述
本关任务：编写一个程序，处理乒乓球联赛的参赛记录。每条参赛记录包含学生的姓名和本次参赛获得的积分。程序需要将所有参赛记录按照总积分从高到低排序输出，如果总积分相同，则按照比赛次数从低到高排序，如果比赛次数也相同，则按照姓名的英文词典顺序排序。

分析
要完成这个任务，我们需要执行以下步骤：

1.读取输入，包括学生的参赛记录。
2.初始化一个数据结构（如字典）来存储每个学生的积分和参赛次数。
3.遍历每条参赛记录，更新学生的积分和参赛次数。
4.对所有学生按照总积分、比赛次数和姓名进行排序。
5.输出排序后的学生信息。

相关知识
为了完成本关任务，你需要掌握：

1.基本的输入输出操作。
2.数据结构，如字典，用于存储学生的积分和参赛次数。
3.列表和元组的操作，用于排序。
4.排序算法，如内置的sorted()函数或者列表的sort()方法，并通过里面的key参数设置排序规则。
5.匿名函数lambda的用法。

数据结构
可以使用字典来存储每个学生的积分和参赛次数，其中键为学生的姓名，值为一个元组，包含积分和参赛次数。

排序算法
使用Python的内置sorted()函数，可以根据多个条件进行排序。

lambda 的用法及示例
在Python中，lambda是一个关键字，用于创建匿名函数，即没有名称的函数。lambda 函数通常用于编写简单的、单行的函数，它们在需要函数对象的地方非常有用，尤其是作为参数传递给另一个函数时。

lambda arguments: expression
# arguments：这是传递给 lambda 函数的参数。
# expression：这是 lambda 函数返回的表达式的结果。
示例：

# 使用lambda函数计算两个数的和
add = lambda x, y: x + y
print(add(5, 3))  # 输出：8
# 使用lambda函数创建一个简单的排序函数
sort_by_second_element = lambda x: x[1] # 根据第二个元素值的大小排序
pairs = [(1, 3), (2, 1), (3, 2)]
sorted_pairs = sorted(pairs, key=sort_by_second_element)
print(sorted_pairs)  # 输出：[(2, 1), (3, 2), (1, 3)]
多条件排序的用法及简单示例
在Python中，sorted() 函数和列表的 sort() 方法可以接受一个 key 参数，该参数是一个函数，用于从每个元素中提取用于排序的值。当需要根据多个条件进行排序时，可以通过 lambda 函数返回一个元组，元组中的每个元素对应一个排序条件。

sorted(iterable, key=lambda x: (condition1, condition2, ...))
# iterable：要排序的可迭代对象。
# key：一个函数，它从每个元素中提取一个用于排序的值。
# condition1, condition2, ...：根据多个条件进行排序。
示例：

# 假设我们有一个学生信息列表，需要根据多个条件进行排序
students = [
    {'name': '张三', 'age': 18, 'score': 98},
    {'name': '李四', 'age': 28, 'score': 95},
    {'name': '王五', 'age': 21, 'score': 98},
]
# 使用sorted函数根据年龄和分数进行排序
sorted_students = sorted(students, key=lambda x: (-x['score'], x['age']))
for student in sorted_students:
    print(student)
在这个示例中，我们首先定义了一个学生信息列表。然后，使用 sorted() 函数和 lambda 函数根据学生的分数和年龄进行排序。lambda 函数返回一个元组，其中包含两个条件：分数（降序）和年龄（升序）。最后，输出排序后的学生信息。

编程要求
根据提示，在右侧编辑器补充代码，实现以下功能：

1.读取学生的参赛记录。
2.存储每个学生的积分和参赛次数。
3.对所有学生按照总积分、比赛次数和姓名进行排序。
4.输出排序后的学生信息。

输入描述
程序的输入包含两部分：

1.第一行包含一个整数n，表示参赛记录的条数。
2.接下来的n行，每行包含两个字符串，第一个是学生的姓名，第二个是本次参赛获得的积分。

输出描述
程序输出排序后的学生信息，格式为"姓名 参赛次数 总积分"，每条记录占一行。

测试说明
平台会对你编写的代码进行测试：

测试输入：
5
Alice 10
Bob 15
Alice 5
Charlie 15
Bob 10
预期输出：
Bob 2 25
Charlie 1 15
Alice 2 15

测试输入：
5
David 20
Eve 10
Frank 20
Eve 5
David 10
预期输出：
David 2 30
Frank 1 20
Eve 2 15
'''

# 读取学生的参赛记录次数
n = int(input())
records = {}

# 遍历每条参赛记录，更新学生的积分和参赛次数
for _ in range(n):
    s = input().split()
    name = s[0]
    score = int(s[1])
    if name in records:
        records[name] = (records[name][0] + 1, records[name][1] + score)
    else:
        records[name] = (1, score)

# 对所有学生按照总积分(降序)、比赛次数(升序)和姓名(升序)进行排序
sorted_records = sorted(records.items(), key=lambda x: (-x[1][1], x[1][0], x[0]))
for name, (count, score) in sorted_records:
    print(name, count, score)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 设置绘图风格
plt.figure(figsize=(10, 6))
ax = plt.gca()

# 定义圆锥母线 (假设 theta = 60度, z = r * tan(theta) 的逆形式，为了方便画 z = 2r 这种陡峭程度)
# 这里为了图示清晰，我们画 z = 1.732 r (大约60度)
theta_deg = 60
theta_rad = np.radians(theta_deg)
r = np.linspace(0, 5, 100)
z = r * np.tan(np.radians(90 - theta_deg)) # 注意题目中theta是与Z轴夹角

# 我们用更简单的比例来示意：z = 2r (即 r = 0.5z)
z_vals = np.linspace(0, 10, 100)
r_vals = 0.5 * z_vals

# 画圆锥侧面线
plt.plot(r_vals, z_vals, 'b-', linewidth=2, label='Cone Surface (True Path)')

# 选取一个微元区间
z_start = 4
z_end = 6
r_start = 0.5 * z_start
r_end = 0.5 * z_end

# 画出 dz (垂直高度)
plt.plot([r_end, r_end], [z_start, z_end], 'r--', linewidth=2, label='dz (Vertical Height)')
# 画出 dr (水平宽度)
plt.plot([r_start, r_end], [z_start, z_start], 'g--', linewidth=2, label='dr (Radius Change)')
# 画出 dl (斜高)
plt.plot([r_start, r_end], [z_start, z_end], 'orange', linewidth=4, alpha=0.6, label='dl (Slant Height)')

# 标注文字
plt.text(r_end + 0.1, (z_start + z_end)/2, 'dz', color='red', fontsize=14, fontweight='bold')
plt.text((r_start + r_end)/2, z_start - 0.3, 'dr', color='green', fontsize=14, fontweight='bold')
plt.text((r_start + r_end)/2 - 0.4, (z_start + z_end)/2 + 0.2, 'dl', color='orange', fontsize=14, fontweight='bold')

# 画出角度 theta (与z轴夹角)
# 在这个视角下，z轴是垂直的。题目定义的theta是与z轴夹角。
# 画一条辅助垂直线
plt.plot([r_start, r_start], [z_start, z_end + 2], 'k:', alpha=0.3)
# 标注角度
plt.text(r_start + 0.1, z_start + 1, r'$\theta$', fontsize=16)

# 模拟"阶梯逼近" (用dz累积)
steps = 5
z_steps = np.linspace(z_start, z_end, steps + 1)
r_steps = np.linspace(r_start, r_end, steps + 1)
for i in range(steps):
    # 画垂直线 (dz)
    plt.plot([r_steps[i+1], r_steps[i+1]], [z_steps[i], z_steps[i+1]], 'r-', alpha=0.5)
    # 画水平线 (dr) - 虽然用户只用了dz，但几何上它是阶梯
    plt.plot([r_steps[i], r_steps[i+1]], [z_steps[i], z_steps[i]], 'g-', alpha=0.5)

plt.title('Why dz != dl: The Geometry of Surface Element', fontsize=16)
plt.xlabel('r (Radius)', fontsize=12)
plt.ylabel('z (Height)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.axis('equal') # 保持比例一致，否则看不出长度区别

# 添加说明文字
plt.text(1.5, 8, r'$dl = \sqrt{dz^2 + dr^2} = \frac{dz}{\cos\theta}$', fontsize=14, bbox=dict(facecolor='white', alpha=0.8))
plt.text(1.5, 6.5, r'Area via $dl$: $2\pi r \cdot dl$ (Correct)', fontsize=12, color='orange')
plt.text(1.5, 5.8, r'Area via $dz$: $2\pi r \cdot dz$ (Wrong)', fontsize=12, color='red')

plt.tight_layout()
plt.savefig('dz_vs_dl_comparison.png')
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
def mian():
    num = int(input("请输入一个非负整数："))
    if num < 0:
        print("输入错误，请输入非负整数。")
    else:
        result = factorial(num)
        print(f"{num} 的阶乘是 {result}")

if __name__ == "__main__":
    mian()
```
