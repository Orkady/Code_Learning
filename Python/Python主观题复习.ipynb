2. 请描述以下程序的功能，并写出下面程序的执行结果

````python
def f(seq):
    if type(seq) == dict:
        seq = list(seq.values())
    else:
        seq = list(seq)
    x = seq[0]
    for i in seq:
        if i > x:
            x = i
    return x
    
print(f([1, 2, 4, 5, 7, 9, 10]))
````

**功能：** 该函数接受一个变量 `seq` ，先进行类型判断，如果为字典，将字典的值取出存在列表中，如果是其他类型，则直接转换为列表。之后取列表 `seq` 的第一个值为 `x` 的初始值，遍历转换的列表中的元素，每个元素与 `x` 值进行比较，若大于 `x` 值，则替换该 `x` 为新的 `x` 值，直到遍历完整个列表序列，返回最终的 `x` 值，该 `x` 即为整个列表序列中的最大值

6. 请描述以下程序的功能，并写出下面程序的执行结果 -> ⭐

````python
def func(li, num):
    if num == 1:
        li.append(1)
        return 1
    elif num >= 2:
        li.append(num * func(li, num - 1))
        return li[-1] + li[-2]
        
mylist = []
result = func(mylist, 4)

print(mylist, result, sep=",")
````

**功能：** 
该程序定义了一个递归函数 `func(li, num)` ，其功能是生成一个特定序列并返回最后两个元素的和。具体行为如下：当 `num` 为 1 时，向列表 `li` 添加元素 1，并返回 1。当 `num` >= 2 时，递归调用 `func(li, num-1)` ，然后将 `num` 乘以递归结果的值添加到 `li` 中，最后返回列表 `li` 中最后两个元素的和。

**执行结果：**

1. 编写程序，输入一个数，判断这个数是不是素数（一个大于1的正整数，如果除了1和它本身以外，不能被其他正整数整除，就叫素数）

````python
# 输入一个正整数
num = int(input("请输入一个正整数: "))

# 判断是否为素数
if num <= 1:
    # 小于等于1的数不是素数
    print(f"{num}不是素数")
else:
    # 假设是素数，然后检查是否能被整除
    is_prime = True
    
    # 从2开始到num-1，判断是否能整除num
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break  # 找到一个因数就不是素数，结束循环
    
    # 输出结果
    if is_prime:
        print(f"{num}是素数")
    else:
        print(f"{num}不是素数")
````

2. 设计一个Python函数 `check_rose_number()` ，该函数接受一个四位数作为参数，并返回一个字符串，指示该数是否为玫瑰花数。玫瑰花数定义为：一个数其各位数字的四次方之和等于该数本身

````python
def check_rose_number(num):
    # 将数字转换为字符串，方便获取各位数字
    str_num = str(num)
    
    # 计算各位数字的四次方之和
    sum_pow = 0
    for digit in str_num:
        sum_pow += int(digit) ** 4  # 四次方
    
    # 判断是否为玫瑰花数
    if sum_pow == num:
        return "是玫瑰花数"
    else:
        return "不是玫瑰花数"

# 测试示例
'''唐：这个考试时候不用写，写上面那一点点就可以了，放在这里看下效果'''
print(check_rose_number(1634))
print(check_rose_number(8208))
print(check_rose_number(1234))
````

3. 实现一个函数 `generate_narcissistic_numbers()` ，该函数将生成并返回一个包含所有水仙花数的列表。水仙花数定义为：一个数其各位数字的三次方之和等于该数本身

````python
'''唐：水仙花数一般指的三位数，这里题目没有指明，可能是想要考察代码逻辑，就按不知道水仙花数是三位数来处理，如果考试题目提到了水仙花数为三位数，就在range()这里改为1000'''
def generate_narcissistic_numbers():
    # 存储结果的列表
    result = []
    
    # 检查从0到9999的所有数
    '''唐：如果考试提到了水仙花数是三位数，修改range()的范围，或者使用更简单的逻辑：直接提取百位、十位、个位的数字，进行三次方并相加比较'''
    for num in range(0, 10000):
        # 计算各位数字的三次方之和
        sum_cubes = 0
        temp = num  # 临时变量用于分解数字
        while temp > 0:
            digit = temp % 10      # 获取个位数
            sum_cubes += digit ** 3  # 累加三次方
            temp //= 10            # 去掉个位
        
        # 注意：num为0时，循环不执行，sum_cubes=0，相等
        
        # 判断是否满足水仙花数条件
        if sum_cubes == num:
            result.append(num)
    
    return result

# 调用函数并打印结果
print(generate_narcissistic_numbers())
````

4. “百马百瓦问题”：有100匹马驮100块瓦，大马驮3块，小马驮2块，两个马驹驮1块。问大马、小马、马驹各有多少匹？

````python
# 遍历所有可能的大马数量
for x in range(0, 101):  # 大马从0到100
    # 遍历所有可能的小马数量
    for y in range(0, 101 - x):  # 小马从0到100-x
        z = 100 - x - y  # 马驹数量
        # 检查瓦数条件（使用整数运算避免浮点误差）
        if 6 * x + 4 * y + z == 200:  # 3x+2y+0.5z=100 乘以2得6x+4y+z=200
            print(f"大马: {x}匹, 小马: {y}匹, 马驹: {z}匹")
````

5. 输出一个乘法表。要求输入一个整数n(n限定在1-9之间)，输出n*n的乘法表，乘法表打印出来为下三角样式，格式工整

````python
# 输入整数n（1-9）
n = int(input("请输入一个1-9之间的整数: "))

# 打印n*n乘法表（下三角样式）
for i in range(1, n + 1):      # 外层循环控制行数（乘数）
    for j in range(1, i + 1):  # 内层循环控制列数（被乘数，只到i为止）
        # 计算乘积
        result = i * j
        # 格式化输出，每个等式占4个字符宽度
        '''唐：这里可以不用使用{result:2}这么规整美观，也可以使用\t制表符来进行分割'''
        print(f"{j}×{i}={result:2}", end="  ")
    print()  # 每行结束后换行
````

6. 编写程序，计算并输出100到500（不包括500）内所有能被2整除但不能被3整除的数的和

````python
# 初始化总和变量
total = 0

# 遍历100到499（不包括500）
for num in range(100, 500):
    # 判断条件：能被2整除且不能被3整除
    if num % 2 == 0 and num % 3 != 0:
        total += num  # 累加符合条件的数

# 输出结果
print(f"100到500内所有能被2整除但不能被3整除的数的和为: {total}")
````

8. 编写一个Python函数 `calculate_ticket_price()` ，该函数接受用户年龄作为参数，并返回相应的票价。成人票价为20美元，青少年（13-18岁）为10美元，儿童（3-12岁）为5美元，3岁以下免费，65岁以上老人半价（10美元）

````python
def calculate_ticket_price(age):
    # 根据年龄计算票价
    if age < 3:
        return 0      # 3岁以下免费
    elif age <= 12:
        return 5      # 儿童票5美元
    elif age <= 18:
        return 10     # 青少年票10美元
    elif age < 65:
        return 20     # 成人票20美元
    else:
        return 10     # 65岁以上老人半价（10美元）

# 测试示例
'''唐：考试可以不用写如下'''
print(calculate_ticket_price(2))   # 0
print(calculate_ticket_price(5))   # 5
print(calculate_ticket_price(15))  # 10
print(calculate_ticket_price(30))  # 20
print(calculate_ticket_price(70))  # 10
````

9. 编写程序，定义一个工程师类，类名为 `Engineer` 。
成员变量包含 `e_id` 、 `name` 、 `Rank` ，用来表示工程师的工号、姓名和职级。其中， `e_id` 和 `name` 接收实例化参数， `Rank` 初值为1。
成员函数包括 `print_info()` 函数，执行时输出工程师信息； `up_Rank()` 函数，执行时将工程师职级提升一个级别； `down_Rank()` 函数，执行时将工程师职级下降一个级别。注意：职级不能小于1。
类定义后，实例化一个对象： `e1` ，工号为"111"，姓名为"李四"。
若马上执行 `e1` 的 `print_info()` 方法，输出形式为： `engineer:111,name:李四,Rank:1` ；若执行 `e1` 的 `up_Rank()` 方法，输出 `engineer:111's Rank+1,当前Rank:2` ；若马上执行 `e1` 的 `down_Rank()` 方法，输出 `engineer:111's Rank-1,当前Rank:0`

````python
class Engineer:
    def __init__(self, e_id, name):
        self.e_id = e_id      # 工号
        self.name = name      # 姓名
        self.Rank = 1         # 职级，初值为1
    
    def print_info(self):
        # 输出工程师信息
        print(f"engineer:{self.e_id},name:{self.name},Rank:{self.Rank}")
    
    def up_Rank(self):
        # 提升职级
        self.Rank += 1
        print(f"engineer:{self.e_id}'s Rank+1,当前Rank:{self.Rank}")
    
    def down_Rank(self):
        # 降低职级，但职级不能小于1
        if self.Rank > 1:
            self.Rank -= 1
            print(f"engineer:{self.e_id}'s Rank-1,当前Rank:{self.Rank}")
        else:
            # 如果职级已经是1，则不能再降
            print(f"engineer:{self.e_id}的Rank已经是1，不能再降")

# 实例化一个对象
e1 = Engineer("111", "李四")

# 测试
e1.print_info()      # 输出：engineer:111,name:李四,Rank:1
e1.up_Rank()         # 输出：engineer:111's Rank+1,当前Rank:2
e1.down_Rank()       # 输出：engineer:111's Rank-1,当前Rank:1
e1.down_Rank()       # 输出：engineer:111的Rank已经是1，不能再降
````
