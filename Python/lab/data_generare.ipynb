```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_and_save_data(filename='function_data.csv', num_samples=2000):
    """
    生成符合指定函数分布的数据并保存为CSV
    函数: y = 5*sin(10*x) + 12*cos(5*x) + 10*(sin(12*x)^3) + 6*noise1 + 6*noise2
    """
    # 1. 设置随机种子，保证每次生成的数据一致（方便复现）
    np.random.seed(2025)

    # 2. 生成 x 数据 (在 -3 到 3 之间均匀采样，区间可根据需要调整)
    x = np.linspace(-10, 10, num_samples)

    # 3. 生成两个独立的噪声源 (服从标准正态分布 N(0, 1))
    noise1 = np.random.normal(0, 1, num_samples)
    noise2 = np.random.normal(0, 1, num_samples)

    # 4. 计算 y 值
    y = 5 * np.sin(10 * x) + \
        12 * np.cos(5 * x) + \
        10 * (np.sin(12 * x)**3) + \
        2 * np.log(11+ x) + \
        6 * noise1 + \
        3 * noise2

    # 5. 保存为 DataFrame 并导出 CSV
    df = pd.DataFrame({'x': x, 'y': y})
    df.to_csv(filename, index=False)
    
    print(f"✅ 数据已成功生成并保存至: {filename}")
    print(f"数据量: {num_samples} 条")
    print("-" * 30)
    print("前5行数据预览：")
    print(df.head())

    # 6. (可选) 可视化预览，确认难度是否合适
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, s=5, alpha=0.5, label='Noisy Data')
    
    # 绘制无噪声的真实曲线供对比
    y_true = 5 * np.sin(10 * x) + 12 * np.cos(5 * x) + 10 * (np.sin(12 * x)**3)
    plt.plot(x, y_true, color='red', linewidth=2, label='True Function (No Noise)')
    
    plt.title("Data Distribution Preview")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    generate_and_save_data()
```
